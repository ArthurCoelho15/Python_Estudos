# Ideia 

```python
movies = ["Tumulo dos vagalumes", "flow", "castelo encantado", "soul"]
num_oscars = [5, 2, 1, 3]

# tornando os nomes dos filmes em capitalize (primeira letra maiscula)
for filme in range(len(movies)):
    movies[filme] = movies[filme].title()

# compacta as duas listas utilizando zip com o sorted de *args e depois descompacta para utilizar nos eixos,
# key=lambda x: x[1] ta sendo utilizado para ordenar pelo oscars
sorted_movies, sorted_oscars = zip(*sorted(zip(movies, num_oscars), key=lambda x: x[1]))

# conferindo a ordenao
print(sorted_movies)
print(sorted_oscars)
```

```python
# definindo um ttulo para o grfico
plt.title("Meus filmes favoritos")

# plt.bar(numero de barras, nomes de cada uma, largura da barra, borda, espessura_da_borda)
bars = plt.bar(range(len(sorted_movies)), sorted_oscars, width=1, edgecolor="black", linewidth=0.8)

# plt.axis(inicio_x, fim_x, inicio_y, fim_y)
plt.axis([-0.7, len(sorted_movies)-0.3, 0, 6])
#plt.grid(True)

# plt.xlabel / plt.ylabel  => Ttulo dos eixos
plt.ylabel("Numero de premiaes")
plt.xlabel("Filmes")

# plt.bar_label(quais barras rotular, tipo de dado (%d inteiro; %f float; %s string), distanciamento da barra) => Data label in bar
plt.bar_label(bars, fmt='%d', padding=5)

# plt.xticks(quantidade de barras, nome do parametro a ser utilizado)
plt.xticks(range(len(sorted_movies)), sorted_movies)

# imprime o grfico
plt.show()
```

# Gemini para ajudar

```python
movies = ["Tumulo dos vagalumes", "flow", "castelo encantado", "soul"]
num_oscars = [5, 2, 1, 3]

sorted_movies, sorted_oscars = zip(*sorted(zip(movies, num_oscars), key=lambda x: x[1]))

plt.title("Lista de filmes")
bars = plt.bar(range(len(sorted_movies)), sorted_oscars, color="lightgreen")

for bar in bars:
    yval = bar.get_height() / 2  # Posio central
    plt.text(bar.get_x() + bar.get_width()/2, yval, str(bar.get_height()),
             ha='center', va='center', color='black', fontsize=12)

plt.ylabel("Numero de premiaes")
plt.xlabel("Filmes")
plt.xticks(range(len(sorted_movies)), sorted_movies)

plt.show()
```
